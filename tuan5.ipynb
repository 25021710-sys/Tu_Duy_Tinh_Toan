```python
def max_of_two(a,b):
    if a>=b:
        return a
    else:
        return b
max_of_two(3,2)
```

```python
def swap (a,b):
    a,b=b,a
    return a,b
a,b=map(int,input().split())
print (swap(a,b))
```

```python
def snt(a):
    k=0
    for i in range (1,a):
        if a%i==0:
            k+=1
    if k==1: return "True"
    else: return "False"
snt(5)
```

```python
def vi_tri(A,a):
    for i in range(len(A)):
        if A[i]==a: 
            return i+1
    return "-1"
A=list(map(int, input().split()))
a=int(input())
print(vi_tri(A,a))
```

```python
def shh(a):
    k=0
    for i in range (1,a):
        if a%i==0:
            k+=i
    if a==k: return "true"
    else: return "false"
shh(5)
    
```

```python
def giai_thua(a):
    p=1
    for i in range(1,a+1):
        p*=i
    return p
n=int(input())
print(giai_thua(n))
```

```python
def tinh_toan(a,b,c):
    if c=="+":
        return a+b
    if c=="-":
        return a-b
    if c=="*":
        return a*b
    if c=="/":
        return a/b
a,b=map(int,input().split())
c=input()
print(tinh_toan(a,b,c))

```

```python
def hamming(x, y):
    xor_value = x ^ y
    count = 0
    while xor_value > 0:
        count += xor_value & 1
        xor_value >>= 1
    
    return count
hamming(1, 4)
```

```python
def tong_chu_so(n):
    return sum(int(ch) for ch in str(n))
tong_chu_so(1234)
```

```python
s1, s2 = input().split()
def dang_cau(s1, s2):
    M = []
    for i in range(26):
        M.append(0)
    for i in range(len(s1)):
        if M[ord(s1[i]) - 97] != 0:
            if M[ord(s1[i]) - 97] != s2[i]:
                return "false"
        else:
            M[ord(s1[i]) - 97] = s2[i]
    return "true"
print(dang_cau(s1, s2))
```
