```python
def collazt(n):
    k=1
    while n!=1:
        k+=1
        if n%2==0:
            n=n/2
        else: n=3*n+1
    return k
n=int(input("n="))
A=[]
B=[]
for i in range (1,n+1):
    A.append(collazt(i))
for j in range (1,n+1):
    if max(A)==collazt(j):
        B.append(j)
print(min(B),max(A))
```

```python
s = input("Nhập chuỗi: ")
count = 0
in_word = False

for i in range(0, s.__len__()):
    if s[i] != ' ':
        if not in_word:
            count += 1
            in_word = True
    else:
        in_word = False

print("Số từ trong chuỗi là:", count)
```

```python
s = input("Nhập chuỗi: ")
upper = 0
lower = 0
digit = 0

for c in s:
    if c.isupper():
        upper += 1
    elif c.islower():
        lower += 1
    elif c.isdigit():
        digit += 1

print("Chữ hoa:", upper)
print("Chữ thường:", lower)
print("Chữ số:", digit)
```

```python
s = input("Nhập mật khẩu: ")
has_upper = False
has_lower = False
has_digit = False
has_special = False

for c in s:
    if c.isupper():
        has_upper = True
    elif c.islower():
        has_lower = True
    elif c.isdigit():
        has_digit = True
    elif not c.isalnum():
        has_special = True

if has_upper and has_lower and has_digit and has_special and s.__len__() > 6:
    print("Mật khẩu mạnh")
else:
    print("Mật khẩu yếu")
```

```python
s = input("Nhập số: ")
s = s[::-1]  
result = ""

for i in range(0, s.__len__()):
    result += s[i]
    if (i + 1) % 3 == 0 and i != s.__len__() - 1:
        result += '.'

print("Kết quả:", result[::-1])
```

```python
a = input("Nhập chuỗi a: ")
b = input("Nhập chuỗi b: ")
start = -1
for i in range(len(a) - len(b) + 1):
    match = True
    for j in range(len(b)):
        if a[i + j] != b[j]:
            match = False
            break
    if match:
        start = i
        break
if start != -1:
    new_a = ""
    for i in range(len(a)):
        if i < start or i >= start + len(b):
            new_a += a[i]
    print("Chuỗi sau khi xóa là:", new_a)
else:
    print("Không tìm thấy chuỗi b trong chuỗi a.")
```
