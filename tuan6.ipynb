```python
line = input().strip()
pairs = line.split()
result = {}
for pair in pairs:
    if ':' in pair:  
        key, value = pair.split(':')  
        if key in result:
            result[key].append(value)
        else:
            result[key] = [value] 
print(result)
```

```python
s=list(map(int, input().split()))
result={}
a,b,c=0,0,0
for i in range (len(s)):
    if s[i]>0:
        a+=1
    elif s[i]==0:
        b+=1
    else:
        c+=1
result['positives']=a
result['negatives']=c
result['zeros']=b
print (result)
```

```python
s1=eval(input())
s2=eval(input())
result={}
all_keys= set(s1.keys()) | set(s2.keys())
for key in all_keys:
    result[key]=s1.get(key)+s2.get(key)

print(dict(sorted(result.items())))
```

```python
s=list(map(int, input().split()))
n= int(input())
M=[]
for i in range (len(s)):
    for j in range(i+1,len(s)):
        if s[i]+s[j]==n:
            t=(s[i],s[j])
            M.append(t)
list(set(M))
M.sort()           
print(M)   
```

```python
def vi_tri(n,k):
    for i in range (len(n)):
        if n[i]==k:
            return i+1 
    return "-1"
n=list(map(int,input().split()))
k=int(input())
print(vi_tri(n,k))
```
